# Demonstration of HxRG Simulator

```python
%matplotlib notebook
from hxrg_simulator import detector 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the detector
H2RG = detector.detector()  # Default is JWST NIRSpec characteristics
```

```python
# Load a sample flux image
FluxImage = np.load('HPFLogo2k2k.npy')
plt.imshow(FluxImage)
```

```python
# Expose the detector for 30 seconds using this flux image
DataCube = H2RG.take_exposure(FluxImage,itime=30)
```

```python
print(DataCube.shape)
# Second readout
plt.imshow(DataCube[1,:,:])
```
